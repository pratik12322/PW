### 1. Difference between a Function and a Method in Python
**Function:** A block of reusable code, defined using `def`, that can be called independently.

```python
def greet():
    print("Hello")

greet()
```

**Method:** A function that is associated with an object, typically defined in a class. `say_hi()` is a method because it's tied to an object of the class `Person`.

```python
class Person:
    def say_hi(self):
        print("Hi")

p = Person()
p.say_hi()
```

### 2. Function Arguments and Parameters in Python
**Parameter:** Variable listed inside the parentheses when defining a function.

**Argument:** Actual value passed when calling the function.

```python
def add(a, b):   # a and b are parameters
    return a + b

add(5, 10)       # 5 and 10 are arguments
```

### 4. Purpose of the `return` Statement
The `return` statement sends a result back to the caller and exits the function.

```python
def multiply(x, y):
    return x * y

result = multiply(2, 3)
print(result)  # result is 6
```

### 5. Iterators vs Iterables
**Iterable:** Any object you can loop over (like a list, string, tuple).

**Iterator:** An object that produces values one at a time using `__next__()`.

```python
nums = [1, 2, 3]         # Iterable
it = iter(nums)          # Now it's an iterator
print(next(it))          # Output: 1
```

### 6. Generators and How They Are Defined
A generator is a special function that `yields` values one by one, saving state between calls.

Defined using `yield` instead of `return`.

```python
def count_up_to(n):
    i = 1
    while i <= n:
        yield i
        i += 1

for number in count_up_to(5):
    print(number)
```

### 7. Advantages of Generators
* **Memory efficient:** Doesnâ€™t store all values in memory.
* **Lazy evaluation:** Computes values as needed.
* **Ideal for large data processing,** like reading big files or streams.

### 8. Lambda Functions
A lambda is an anonymous, one-line function.
Used when: You need a short function temporarily, often with `map()`, `filter()`, or `sorted()`.

```python
double = lambda x: x * 2
print(double(5))
```

### 9. Purpose and Usage of map()
`map()` applies a function to every item in an iterable.

```python
nums = [1, 2, 3, 4]
result = map(lambda x: x * 2, nums)
print(list(result))
```

### 10. Difference Between `map()`, `reduce()`, and `filter()`
*   **`map(func, iterable)`**: Transforms each element.
*   **`filter(func, iterable)`**: Filters elements where function returns `True`.
*   **`reduce(func, iterable)`** (from `functools`): Reduces the iterable to a single value by applying the function cumulatively.

```python
from functools import reduce

# Example list
nums = [1, 2, 3, 4]

# map
print("map:", list(map(lambda x: x*2, nums)))

# filter
print("filter:", list(filter(lambda x: x % 2 == 0, nums)))

# reduce
print("reduce:", reduce(lambda x, y: x + y, nums))
```

```python
def sum_even_numbers(numbers):
    return sum(num for num in numbers if num % 2 == 0)

# Example
print(sum_even_numbers([1, 2, 3, 4, 5, 6]))
```

```python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

# Check primes from 1 to 200
primes = [x for x in range(1, 201) if is_prime(x)]
print(primes)
```

```python
class Fibonacci:
    def __init__(self, terms):
        self.terms = terms
        self.a, self.b = 0, 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.terms:
            raise StopIteration
        self.count += 1
        self.a, self.b = self.b, self.a + self.b
        return self.a

# Example
for num in Fibonacci(10):
    print(num, end=" ")
```

```python
def powers_of_two(n):
    for i in range(n + 1):
        yield 2 ** i

# Example
for val in powers_of_two(5):
    print(val)
```

```python
def remove_vowels(s):
    vowels = 'aeiouAEIOU'
    return ''.join(filter(lambda ch: ch not in vowels, s))

# Example
print(remove_vowels("Hello World"))
```

You are given a list of book orders. Each order is a list with the following structure:

`[Order Number, Book Title & Author, Quantity, Price per Item]`

Your task is to process these orders and return a list of tuples, where each tuple contains the `Order Number` and the `Final Order Value`.

The `Final Order Value` is calculated as `Quantity * Price per Item`. However, if the `Final Order Value` is less than 100.00, a small order charge of 10 euros should be added.

**Use `lambda` and `map()` to solve this problem.**

```python
orders = [
    [34587, "Learning Python, Mark Lutz", 4, 40.95],
    [98762, "Programming Python, Mark Lutz", 5, 56.80],
    [77226, "Head First Python, Paul Barry", 3, 32.95],
    [88112, "EinfÃ¼hrung in Python3, Bernd Klein", 3, 24.99]
]

result = list(map(
    lambda order: (order[0], order[2] * order[3] if order[2] * order[3] >= 100 else order[2] * order[3] + 10),
    orders
))

print(result)
```
