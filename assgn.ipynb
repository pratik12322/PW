> **Question 1: What is the difference between descriptive statistics
> and inferential‬**
>
> **statistics? Explain with examples.‬**
>
> **Ans -‬**
>
> **Difference between Descriptive and Inferential Statistics‬**
>
> **1. Descriptive Statistics:‬**
>
> **●‬ ‭Descriptive statistics is the branch of statistics that deals with
> organizing,‬**
>
> **summarizing, and presenting data in a meaningful way.‬**

**●‬ ‭It does not make any predictions or generalizations beyond the data
collected.‬**

**●‬ ‭Common techniques include mean, median, mode, standard deviation,
charts,‬**

> **graphs, and tables.‬**
>
> **●‬ ‭Example: Suppose we collect the marks of 100 students in a college
> and find:‬**
>
> **○‬ ‭Mean = 72‬**
>
> **○‬ ‭Median = 70‬**
>
> **○‬ ‭Mode = 68‬**
>
> **○‬ ‭Standard Deviation = 5‬**
>
> **This is descriptive statistics because it simply describes the
> performance‬**
>
> **of these 100 students without making any claim about other
> students.‬**
>
> **2. Inferential Statistics:‬**
>
> **●‬ ‭Inferential statistics helps us to draw conclusions, make
> predictions, or test‬**
>
> **hypotheses about a population based on a sample of data.‬**
>
> **●‬ ‭It uses probability theory to make these inferences.‬**
>
> **●‬ ‭Common techniques include hypothesis testing, confidence
> intervals, correlation,‬** **regression, t-tests, and ANOVA.‬**
>
> **●‬ ‭Example: From a sample of 100 students in a college, we conclude
> with 95%‬confidence that the average marks of all students in the
> college is between 70–74.‬Here, we are inferring about the whole
> population using the sample.‬**
>
> **Question 2: What is sampling in statistics? Explain the differences
> between random and‬stratified sampling.‬**
>
> **Ans -‬**
>
> **Sampling is the process of selecting a subset (sample) from a larger
> group (population)‬to collect data, analyze it, and make conclusions
> about the entire population.‬**  
> **Since studying the entire population is often time-consuming,
> expensive, or impractical,‬sampling makes research faster, easier, and
> more cost-effective.‬**
>
> **●‬ ‭Example: If a university has 10,000 students, and you survey only
> 500 students to‬find the average study hours, that 500 is your sample,
> and the 10,000 is your‬population.‬**
>
> **Types of Sampling: Random vs Stratified Sampling‬**
>
> **1. Random Sampling:‬**
>
> **●‬ ‭Every individual in the population has an equal chance of being
> selected.‬**
>
> **●‬ ‭It is simple and reduces selection bias.‬**
>
> **●‬ ‭Methods include lottery method, using random number tables, or‬**
> **computer-generated random numbers.‬**
>
> **●‬ ‭Example: Selecting 500 students randomly from 10,000 by assigning
> numbers and‬** **using a random number generator.‬**
>
> **Advantages:‬**
>
> **●‬ ‭Easy to perform.‬**
>
> **●‬ ‭Unbiased representation (if sample size is large).‬**
>
> **Disadvantages:‬**
>
> **●‬ ‭May not represent all subgroups (e.g., if most randomly picked
> students are from‬**
>
> **one department).‬**
>
> **2. Stratified Sampling:‬**
>
> **●‬ ‭The population is divided into subgroups (strata) based on certain
> characteristics‬**
>
> **(e.g., gender, department, age group), and samples are taken from
> each subgroup‬**
>
> **proportionally or equally.‬**
>
> **●‬ ‭Ensures all subgroups are represented.‬**
>
> **●‬ ‭Example: Dividing the 10,000 students by department (CSE, ECE,
> Mechanical,‬**
>
> **Civil) and then selecting students proportionally from each
> department.‬**
>
> **Advantages:‬**
>
> **●‬ ‭More accurate and representative.‬**
>
> **●‬ ‭Reduces sampling error.‬**
>
> **Disadvantages:‬**
>
> **●‬ ‭Requires prior knowledge of population structure.‬**
>
> **●‬ ‭More complex than random sampling.‬**
>
> **Question 3: Define mean, median, and mode. Explain why these
> measures of central‬**
>
> **tendency are important.‬**
>
> **Ans -‬**
>
> **Mean, Median, and Mode – Measures of Central Tendency‬**
>
> **Central tendency refers to statistical measures that describe the
> center or typical value‬**
>
> **of a dataset. The three most common measures are Mean, Median, and
> Mode.‬**
>
> **1. Mean (Arithmetic Mean)‬**
>
> **●‬ ‭Mean is the sum of all observations divided by the total number of
> observations.‬**
>
> **●‬ ‭Formula:‬**
>
> **Mean=Σxn\text{Mean} = \frac{\Sigma x}{n}Mean=nΣx​**

**where Σx\Sigma xΣx = sum of all values, and nnn = total number of
values.‬**

> **●‬ ‭Example: If marks = 10, 20, 30,‬**
>
> **Mean = (10 + 20 + 30) ÷ 3 = 60 ÷ 3 = 20.‬**
>
> **●‬ ‭Use: Best for normally distributed data.‬**
>
> **2. Median‬**
>
> **●‬ ‭Median is the middle value when data is arranged in ascending or
> descending‬**
>
> **order.‬**
>
> **●‬ ‭If the number of observations is odd → middle value is the
> median.‬**
>
> **●‬ ‭If even → median is the average of the two middle values.‬**
>
> **●‬ ‭Example: Data = 5, 10, 15, 20, 25‬**
>
> **Median = 15 (middle value).‬**

> **3. Mode‬**
>
> **●‬ ‭Mode is the value that occurs most frequently in the data.‬**
>
> **●‬ ‭A dataset may have no mode, one mode (unimodal), or multiple
> modes‬**
>
> **(bimodal/multimodal).‬**
>
> **●‬ ‭Example: Data = 2, 4, 4, 6, 8‬**
>
> **Mode = 4 (appears most often).‬**
>
> **●‬ ‭Use: Useful for categorical data (e.g., finding the most common
> shoe size).‬**
>
> **Importance of Mean, Median, and Mode‬**
>
> **●‬ ‭They summarize large datasets with a single representative
> value.‬**
>
> **●‬ ‭Help in data comparison between different groups or time
> periods.‬**
>
> **●‬ ‭Used in decision making (business, economics, research,
> healthcare).‬**
>
> **●‬ ‭Each measure has its specific application:‬**
>
> **○‬ ‭Mean → Best for balanced data.‬**
>
> **○‬ ‭Median → Best when there are outliers.‬**
>
> **○‬ ‭Mode → Best for categorical or discrete data.‬**

> **Question 4: Explain skewness and kurtosis. What does a positive skew
> imply about the‬**
>
> **data?‬**
>
> **Ans -‬**
>
> **Skewness and Kurtosis‬**
>
> **Both are measures that describe the shape and distribution of
> data.‬**
>
> **1. Skewness‬**
>
> **●‬ ‭Skewness measures the asymmetry of a data distribution around its
> mean.‬**
>
> **●‬ ‭A perfectly symmetrical distribution (like the normal
> distribution) has a skewness‬**
>
> **= 0.‬**
>
> **Types of Skewness:‬**
>
> **●‬ ‭Positive Skew (Right Skew):‬**
>
> **○‬ ‭Tail is stretched to the right side (higher values).‬**
>
> **○‬ ‭Mean \> Median \> Mode.‬**
>
> **○‬ ‭Example: Income distribution (few people earn very high
> salaries).‬**
>
> **●‬ ‭Negative Skew (Left Skew):‬**
>
> **○‬ ‭Tail is stretched to the left side (lower values).‬**
>
> **○‬ ‭Mean \< Median \< Mode.‬**
>
> **○‬ ‭Example: Retirement age (most people retire around a similar age,
> few‬**
>
> **much earlier).‬**
>
> **2. Kurtosis‬**
>
> **●‬ ‭Kurtosis measures the "peakedness" or flatness of a distribution
> compared to a‬**
>
> **normal distribution.‬**
>
> **●‬ ‭Types of Kurtosis:‬**
>
> **○‬ ‭Mesokurtic (Kurtosis = 3): Normal distribution.‬**
>
> **○‬ ‭Leptokurtic (Kurtosis \> 3): Sharper peak, fatter tails → more
> outliers.‬**

> **Question 6: Compute the covariance and correlation coefficient
> between the following‬two datasets provided as lists in Python: list_x
> = \[10, 20, 30, 40, 50\] list_y = \[15, 25, 35, 45,‬60\]‬**
>
> **Ans -‬**  
> ***<u>PYTHON CODE :</u>‬***
>
> **import numpy as np‬**  
> **list_x = \[10, 20, 30, 40, 50\]‬**  
> **list_y = \[15, 25, 35, 45, 60\]‬**  
> **x = np.array(list_x)‬**  
> **y = np.array(list_y)‬**  
> **cov_matrix = np.cov(x, y, bias=False)‬**  
> **covariance = cov_matrix\[0, 1\]‬**  
> **correlation = np.corrcoef(x, y)\[0, 1\]‬**  
> **print("List X:", list_x)‬**  
> **print("List Y:", list_y)‬**  
> **print("Covariance:", covariance)‬**  
> **print("Correlation Coefficient:", correlation)‬**
>
> ***<u>OUTPUT :</u>‬***
>
> **List X: \[10, 20, 30, 40, 50\]‬**  
> **List Y: \[15, 25, 35, 45, 60\]‬**  
> **Covariance: 225.0‬**  
> **Correlation Coefficient: 0.9912407071619302‬**

> **Question 8: You are working as a data analyst in an e-commerce
> company. The marketing‬team wants to know if there is a relationship
> between advertising spend and daily sales.‬● Explain how you would use
> covariance and correlation to explore this relationship. ●‬Write Python
> code to compute the correlation between the two lists:
> advertising_spend =‬\[200, 250, 300, 400, 500\] daily_sales = \[2200,
> 2450, 2750, 3200, 4000\]‬**  
> **Ans -‬**
>
> **Exploring Relationship using Covariance and Correlation‬**
>
> **1.‬ ‭Covariance:‬**
>
> **○‬ ‭Covariance measures the direction of the relationship between
> two‬** **variables.‬**
>
> **○‬ ‭If covariance \> 0 → both variables increase together (positive
> relationship).‬**
>
> **○‬ ‭If covariance \< 0 → one increases while the other decreases
> (negative‬** **relationship).‬**
>
> **○‬ ‭If covariance ≈ 0 → no clear relationship.‬**
>
> **○‬ ‭Limitation: It does not show the strength of the relationship.‬**
>
> **2.‬ ‭Correlation:‬**
>
> **○‬ ‭Correlation measures both the direction and strength of the
> relationship.‬**
>
> **○‬ ‭Its value lies between -1 and +1:‬**

**■‬ ‭+1 → Perfect positive relationship‬**

**■‬ ‭-1 → Perfect negative relationship‬**

> **■‬ ‭0 → No linear relationship‬**  
> **○‬ ‭Correlation is a scaled form of covariance, making it easier to
> interpret.‬**
>
> ***<u>PYTHON CODE :</u>‬***  
> **import numpy as np‬**  
> **advertising_spend = \[200, 250, 300, 400, 500\]‬**  
> **daily_sales = \[2200, 2450, 2750, 3200, 4000\]‬**  
> **ad = np.array(advertising_spend)‬**  
> **sales = np.array(daily_sales)‬**  
> **cov_matrix = np.cov(ad, sales, bias=False)‬**  
> **covariance = cov_matrix\[0, 1\]‬**
>
> **correlation = np.corrcoef(ad, sales)\[0, 1\]‬**  
> **print("Advertising Spend:", advertising_spend)‬**  
> **print("Daily Sales:", daily_sales)‬**  
> **print("Covariance:", covariance)‬**  
> **print("Correlation Coefficient:", correlation)‬**
>
> ***<u>OUTPUT :</u>‬***
>
> **Advertising Spend: \[200, 250, 300, 400, 500\]‬**  
> **Daily Sales: \[2200, 2450, 2750, 3200, 4000\]‬**  
> **Covariance: 77500.0‬**  
> **Correlation Coefficient: 0.995871‬**
